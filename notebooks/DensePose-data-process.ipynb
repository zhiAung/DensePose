# （densepose）训练数据集


```python
import json
train_json="/data/xiaobing.wang/peixuan.liang/DensePose-master/detectron/datasets/data/coco/annotations/densepose_coco_2014_train.json"
with open(train_json,"r")as f:
    dict_file=json.load(f)
    #print(type(json_file))
```

```python
random_id=0
for key in dict_file:
    print(key)
print("小训练集的图片数:{}".format(len(dict_file["images"])))
img_list=[]
for key in dict_file["images"][random_id]:
    img_list.append(key)
print("图片的的字典:{}".format(img_list))
print("图片字典举例:{}".format(dict_file["images"][random_id]))

print("小训练集的标注文件数:{}".format(len(dict_file["annotations"])))
#print(dict_file["annotations"][0])
annotions_key_list=[]
for key in dict_file["annotations"][random_id]:
    annotions_key_list.append(key)
print("标注文件的字典:{}".format(annotions_key_list))
for key in dict_file["annotations"][random_id]:
    print("关键字:{}".format(key))
    if (isinstance(dict_file["annotations"][random_id][key],list)):
        print("长度:{}".format(len(dict_file["annotations"][random_id][key])))
    if (isinstance(dict_file["annotations"][random_id][key],int)):
        print("值:{}".format(dict_file["annotations"][random_id][key]))
print("标注文件字典举例:{}".format(dict_file["annotations"][random_id]))
    #print(dict_file[])
```

# (densepose)验证集输出
## body_uv

```python
import cPickle as pickle
val_result_body_uv="/data/xiaobing.wang/peixuan.liang/DensePose-master/test/dense_coco_2014_minival/generalized_rcnn/body_uv_dense_coco_2014_minival_results.pkl"
with open(val_result_body_uv,"r") as f:
    list_file=pickle.load(f)
    
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import numpy as np
random_id=1

root_dir="/data/xiaobing.wang/peixuan.liang/DensePose-master/detectron/datasets/data/coco/val2014"
print(type(list_file))
print("子列表总数为:{}".format(len(list_file)))
result_key_list=[]
for key in list_file[random_id]:
    result_key_list.append(key)
print("body_uv的结果文件的关键字:{}".format(result_key_list))
print("uv数据:{}".format(list_file[random_id][u'uv'].shape))
print("body_uv的结果文件举例:{}".format(list_file[random_id]))
print("")
print("该图片为:")
name_prefix="COCO_val2014_"
name=name_prefix+str(list_file[random_id][u'image_id']).zfill(12)+".jpg"
im=img.imread(os.path.join(root_dir,name))
#im=cv2.imread(os.path.join(root_dir,name))
plt.imshow(im)
plt.show()
print("图片的数据:{}".format(im.shape))


print("visualize the I, U and V images.:")
IUV=list_file[random_id][u'uv']
fig = plt.figure(figsize=[10,10])
plt.imshow(   np.hstack((IUV[0,:,:]/24. ,IUV[1,:,:]/256. ,IUV[2,:,:]/256.))  )#沿着水平方向把数组堆叠起来
plt.title('I, U and V images.')
plt.axis('off') ; plt.show()

print("")
print("visualize the isocontours of the UV fields:")
IUV=list_file[random_id][u'uv']
fig = plt.figure(figsize=[10,10])
plt.imshow( im)#把数组水平翻转
#plt.contour( IUV[:,:,1]/256.,10, linewidths = 1 )
#plt.contour( IUV[:,:,2]/256.,10, linewidths = 1 )#填满等高线的空格
plt.contour( IUV[1,:,:]/256.,10, linewidths = 1 )
plt.contour( IUV[2,:,:]/256.,10, linewidths = 1 )
plt.axis('off') ; plt.show()
```

```python
count=0
count_uv=0
print(IUV.shape)
for i in range(IUV[1,:,1].size):
    for j in range(IUV[1,1,:].size):
        count+=1
        if IUV[1,i,j]!=0:
            count_uv+=1
print("u的总坐标数:{}".format(count))
print("u的有效坐标数:{}".format(count_uv))
```
